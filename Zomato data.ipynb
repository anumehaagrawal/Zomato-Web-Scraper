```python
reviews=[]
for idno in first:
    restaurant=p.getRestaurantReviews(idno)
    rest_details=p.getRestaurantDetails(idno)
    new_dict={}
    arr= restaurant['user_reviews']
    for rev in arr:
        new_dict={}
        new_dict['rating']=rev['review']['rating']
        new_dict['review_text']=rev['review']['review_text']
        new_dict['review_time_friendly']=rev['review']['review_time_friendly']
        new_dict['timestamp']=rev['review']['timestamp']
        new_dict['rest_id']=idno
        new_dict['user_name']=rev['review']['user']['name']
        new_dict['restaurant_city']=rest_details['location']['city']
        new_dict['cuisines']=rest_details['cuisines']
        new_dict['average_cost_for_two']=rest_details['average_cost_for_two']
        new_dict['currency']=rest_details['currency']
        reviews.append(new_dict)
       
        
        
        
    
```
